# Bluestock Mutual Fund Capstone — 10 Key EDA Findings
This notebook summarizes the 10 key findings from the Exploratory Data Analysis (EDA) of mutual fund data.


## 10 Key EDA Findings

**Finding 1:** NAV trends for all 40 schemes mostly rise through the highlighted 2023 bull-run window; see Chart 01.

**Finding 2:** Indexing NAVs to 100 reveals stronger relative return dispersion than absolute NAV plots alone; see Chart 02.

**Finding 3:** SBI Mutual Fund is the clear AUM leader and reaches the annotated ₹12.5 lakh crore dominance level; see Chart 03.

**Finding 4:** The largest AMCs retain a structurally high share of tracked AUM across 2022-2025; see Chart 04.

**Finding 5:** Monthly SIP inflows climb from roughly ₹11.5k crore in January 2022 to the December 2025 high of ₹31,002 crore; see Chart 05.

**Finding 6:** Active SIP accounts and SIP AUM rise together, showing broader participation and larger balances; see Chart 06.

**Finding 7:** The category heatmap shows equity-oriented categories absorbing stronger and more frequent positive net inflows; see Chart 07.

**Finding 8:** SIP ticket-size distributions vary meaningfully by age group, with older groups showing wider amount dispersion; see Chart 10.

**Finding 9:** SIP amount is concentrated in a few leading states while the T30/B30 split shows meaningful non-metro participation; see Charts 13 and 14.

**Finding 10:** Equity fund holdings are materially concentrated in leading sectors, with Banking and IT among the highest aggregate weights; see Charts 18 and 19.
